## Split the data into x and y

```python
x = df[['airline', 'overall', 'traveller_type', 'cabin', 'seat_comfort','cabin_service', 'food_bev', 'entertainment', 'ground_service','value_for_money','Customer_reviews_nlp']]
y = df[['recommended']]
```

## Split the data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=12345)
```

## Build the Model

### 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV

model_lg = LogisticRegressionCV(max_iter=1000)
model_lg.fit(x_train, y_train)
```

```python
# Evaluate the Model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred_lg = model_lg.predict(x_test)
y_true_lg = y_test
accuracy_lg = accuracy_score(y_true_lg, y_pred_lg)
precision_lg = precision_score(y_true_lg, y_pred_lg)
recall_lg = recall_score(y_true_lg, y_pred_lg)
f1_lg = f1_score(y_true_lg, y_pred_lg)

print(f"accuracy = {accuracy_lg}\nprecision = {precision_lg}\nrecall = {recall_lg}\nf1_score = {f1_lg} ")
```

#### Result Visualization

```python
# Visualising the result for airline worth value_for_money
y_pred_lg = model_lg.predict(x_test)
plt.scatter(x_test['airline'][y_pred_lg == 1], x_test['value_for_money'][y_pred_lg == 1], color="red", label="Pass")
plt.scatter(x_test['airline'][y_pred_lg == 0], x_test['value_for_money'][y_pred_lg == 0], color="green", label="Fail")
plt.xlabel('airline')
plt.ylabel('value for money')
plt.title('Visualization of Logisctic regression')
plt.legend()
```
